```python
import pandas as pd
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
nltk.download('punkt') # Download punkt for word_tokenize
nltk.download('wordnet') # Download wordnet for WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize # Import word_tokenize
from nltk.stem import WordNetLemmatizer # Import WordNetLemmatizer
nltk.download('punkt_tab') # Download the punkt_tab resource
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
import joblib
```

```python
df_ml = pd.read_csv('/content/Womens Clothing E-Commerce Reviews.csv')
```

```python
df_ml.drop(['Unnamed: 0', 'Clothing ID'], axis = 1, inplace=True)
```

```python
df_ml.columns= df_ml.columns.str.lower().str.replace('&', '_').str.replace(' ', '_')
```

```python
df_ml = df_ml.rename(columns = {'Review Text' : 'text', 'recommended_ind' : 'recommended',
                          'positive_feedback_count' : 'feedback_count', 'division_name' : 'division',
                          'department_name' : 'department', 'class_name' :'class'})
```

```python
drop_columns = ['age',
                'title',
                'rating',
                'feedback_count',
                'division',
                'department',
                'class']
```

```python
df_ml.drop(drop_columns, axis = 1, inplace = True)
```

```python
df_ml.rename(columns = {'review_text':'text', 'recommended':'recommend'}, inplace = True)
```

```python
df_ml = df_ml.dropna()
```

```python
df_ml["text"].str.isspace().sum()
df_ml[df_ml["text"].str.isspace() == True].index
```

```python
stop_words = stopwords.words('english')
```

```python
def cleaning(data):

    #1. Tokenize
    text_tokens = word_tokenize(data.replace("'", "").lower())

    #2. Remove Puncs
    tokens_without_punc = [w for w in text_tokens if w.isalpha()]

    #3. Removing Stopwords
    tokens_without_sw = [t for t in tokens_without_punc if t not in stop_words]

    #4. lemma
    text_cleaned = [WordNetLemmatizer().lemmatize(t) for t in tokens_without_sw]

    #joining
    return " ".join(text_cleaned)
```

```python
df_ml["text"] = df_ml["text"].apply(cleaning)
```

```python
X = df_ml["text"]
y = df_ml["recommend"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = 101)
```

```python
vectorizer = CountVectorizer()

X_train_count = vectorizer.fit_transform(X_train)
X_test_count = vectorizer.transform(X_test)
```

```python
X_train_count
```

```python
pd.DataFrame(X_train_count.toarray(), columns = vectorizer.get_feature_names_out())
```

```python


logreg_count = LogisticRegression(C = 0.1, max_iter = 1000, class_weight = 'balanced', random_state = 101)
logreg_count.fit(X_train_count,y_train)
```

```python

joblib.dump(logreg_count, 'final_model.pkl')
```

```python
joblib.dump(vectorizer, 'vectorizer.pkl')
```

```python
loaded_model = joblib.load('final_model.pkl')
loaded_vectorizer = joblib.load('vectorizer.pkl')
```

```python
def predict_review(review_text):
    """Predicts the recommendation (0 or 1) for a given review."""
    processed_text = cleaning(review_text)  # Clean the input review

    if not processed_text:  # Handle empty processed text
        print(f"Review: {review_text}")
        print("Prediction: Cannot determine (empty or stopwords-only)")
        return

    text_count = loaded_vectorizer.transform([processed_text])  # Convert text to numerical form
    prediction = loaded_model.predict(text_count)[0]  # Get prediction

    print(f"Review: {review_text}")
    print(f"Prediction: {'Recommended' if prediction == 1 else 'Not Recommended'}")
```

```python
predict_review("This dress is amazing! Fits perfectly.")
predict_review("Terrible quality. Would not recommend.")
predict_review("Dress is very loose.")
predict_review("The material was cheap and it didn't fit well.")
```
